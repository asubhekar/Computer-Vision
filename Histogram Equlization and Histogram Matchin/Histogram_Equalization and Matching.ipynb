## Histogram equalization and Histogram Matching

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
from skimage import exposure
```

### Data statistics

```python
image = cv2.imread('pollen.tif')
image2 = cv2.imread('pollen2.tif')
image3 = cv2.imread('pollen3.tif')
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)
image3 = cv2.cvtColor(image3, cv2.COLOR_BGR2GRAY)

plt.subplot(3,2,1)
plt.title('Original images')
plt. imshow(image, cmap = 'gray')
plt.subplot(3,2,2)
plt.title('Histogram')
plt.hist(image.ravel(), bins = 256, range = [0,256])

plt.subplot(3,2,3)
plt. imshow(image2, cmap = 'gray')
plt.subplot(3,2,4)
plt.hist(image2.ravel(), bins = 256, range = [0,256])


plt.subplot(3,2,5)
plt. imshow(image3, cmap = 'gray')
plt.subplot(3,2,6)
plt.hist(image3.ravel(), bins = 256, range = [0,256])
plt.show()
```

### Histogram Equalizatipon

```python
equalized_image = cv2.equalizeHist(image)
equalized_image2 = cv2.equalizeHist(image)
equalized_image3 = cv2.equalizeHist(image)

plt.subplot(3,2,1)
plt.title('Equalized images')
plt. imshow(equalized_image, cmap = 'gray')
plt.subplot(3,2,2)
plt.title('Histogram')
plt.hist(equalized_image.ravel(), bins = 256, range = [0,256])

plt.subplot(3,2,3)
plt. imshow(equalized_image2, cmap = 'gray')
plt.subplot(3,2,4)
plt.hist(equalized_image2.ravel(), bins = 256, range = [0,256])
plt.show()

plt.subplot(3,2,5)
plt. imshow(equalized_image3, cmap = 'gray')
plt.subplot(3,2,6)
plt.hist(equalized_image3.ravel(), bins = 256, range = [0,256])
plt.show()
```

### Histogram Matching

```python
trees1 = cv2.imread('trees1.jpg')
trees2 = cv2.imread('trees2.jpg')
trees1 = cv2.cvtColor(trees1, cv2.COLOR_BGR2GRAY)
trees2 = cv2.cvtColor(trees2, cv2.COLOR_BGR2GRAY)

plt.subplot(2,2,1)
plt.title('Source')
plt.imshow(trees1)
plt.subplot(2,2,2)
plt.title('Histogram')
plt.hist(trees1.ravel(), bins = 256, range = [0,256])

plt.subplot(2,2,3)
plt.title('Reference')
plt. imshow(trees2)
plt.subplot(2,2,4)
plt.hist(trees2.ravel(), bins = 256, range = [0,256])
plt.show()
```

```python
# histogram matching
multi = True if trees1.shape[-1] > 1 else False
matched = exposure.match_histograms(trees1, trees2, multichannel=multi)

plt.title("Histogram Matched")
plt.axis('off')
plt.imshow(matched)
```
